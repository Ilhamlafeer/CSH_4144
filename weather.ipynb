```python
import pandas as pd
from sklearn import tree
```

```python
data = pd.read_csv(r'weather_forecast.csv')
data.shape
```

```python
data.head(14)
```

```python
X,y = data[['Outlook', 'Temperature', 'Humidity', 'Windy']], data['Play']
```

```python
one_hot_data = pd.get_dummies(data[['Outlook', 'Temperature', 'Humidity', 'Windy', 'Play']], drop_first = False).astype(int)
one_hot_data_X = pd.get_dummies(X, drop_first = False).astype(int)
one_hot_data_y = pd.get_dummies(X, drop_first = True).astype(int)
```

```python
one_hot_data_X.head(14)
```

```python
one_hot_data_y = pd.get_dummies(y, drop_first = True)
one_hot_data_y.head(14)
```

```python
import matplotlib.pyplot as plt
classifier = tree.DecisionTreeClassifier(criterion = 'gini')
classifier = classifier.fit(one_hot_data_X, one_hot_data_y)
plt.figure(figsize=(12,12))
tree.plot_tree(classifier, filled = True,
              feature_names = one_hot_data_X.columns,
               class_names=['No', 'Yes'], fontsize=10
              )
plt.show()
```
